```python
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets

def func(x: float, y: float) -> float:
    """Compute the value of the multivariate function at (x, y)."""
    return x**2 - 4*x + 1 + y**2 - 3*y + 1 + x*y

def gradient_func(x: float, y: float) -> np.array:
    """Compute the gradient of the function at (x, y)."""
    df_dx = 2*x - 4 + y  # Partial derivative with respect to x
    df_dy = 2*y - 3 + x  # Partial derivative with respect to y
    return np.array([df_dx, df_dy])

def build_function():
    x = np.arange(-20., 20., 0.2)
    y = np.arange(-20., 20., 0.2)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)
    return X, Y, Z, func, gradient_func
```

```python
X,Y,Z, f, grad = build_function()
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
plt.show()
```
